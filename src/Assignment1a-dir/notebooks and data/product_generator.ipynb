```python
import numpy as np
import pandas as pd
```

```python

products = [pow(10,2),pow(10,4),pow(10,6),pow(10,8),pow(10,10)]
products = np.asarray(products)
products
min_text_len = np.sqrt(products)
min_text_len
```

```python
product_df = pd.DataFrame(columns=['product','factor1','factor2'])

for product, min_len in zip(products.astype(np.int64),min_text_len.astype(np.int64)):
    for f1 in range(min_len,(product+1)):
        f2 = product / f1
        if (f2.is_integer()):
            new_row = {'product' : product, 'factor1' : f1, 'factor2' : f2}
            product_df = product_df.append(new_row, ignore_index=True)
            print('Product %d: F1: %d, F2: %d' % (product,f1,f2))
    
```

```python
product_df.to_csv('product_combinations.csv',index=False,header=True)
```

```python
product_df = pd.read_csv('product_combinations.csv')
```

```python

product_df['comparison_est'] = product_df['product'] - (product_df['factor2']-1)*product_df['factor2']
product_df.to_csv('product_combinations.csv',index=False,header=True)
```

```python

product_df.head()
```

```python
top_comparisons = product_df.groupby('product')['comparison_est'].nlargest(12)
top_comparisons
```

```python
top_index = []
for i,j in top_comparisons.index.values.tolist():
    top_index.append(j)
top_index
```

```python
top = product_df.loc[top_index]
```
